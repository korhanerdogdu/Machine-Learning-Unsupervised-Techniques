<h2 style="color:rgb(0,120,170)">Task 1 (20 Points): Prove the following Lemma</h2>

Show that for any matrix $X$ and any differentiable scalar-valued function $f$ it holds that 
$$
\frac{\partial f(X^\top X)}{\partial X} = 2{X}\frac{\partial f(X^\top X)}{\partial (X^\top X)}.
$$

########## YOUR SOLUTION HERE ##########


## Solution to Task 1

To prove the lemma, we use the chain rule for matrix calculus. Specifically, we need to find how the matrix \(X\) affects

$$
f(X^\top X)
$$.

### Given:
- \(f\) is a differentiable scalar-valued function.
- \(X\) is any matrix.

### To Prove:
$$
\frac{\partial f(X^\top X)}{\partial X} = 2X\frac{\partial f(X^\top X)}{\partial (X^\top X)}.
$$

### Proof:

1. **Define the function:**
   $$ \begin{aligned} & Y = X^\top X \end{aligned} $$
   
   Then f is a function of Y.

2. **Apply the chain rule:**
   The differential of f with respect to X can be expressed via the chain rule as:
   $$ \begin{aligned} & df = \text{tr} \left( \left(\frac{\partial f}{\partial Y}\right)^\top dY \right) \end{aligned} $$
   where $$ \begin{aligned} & dY = d(X^\top X) = X^\top dX + (dX)^\top X \end{aligned} $$.

3. **Differentiate \( Y \):**
   $$ \begin{aligned} & dY = X^\top dX + (dX)^\top X \end{aligned} $$
   
   Note that the trace of \(A^\top B\) equals the sum of the products of corresponding elements of \(A\) and \(B\), and for any matrix \(A\), the trace of \(A^\top\) is the same as the trace of \(A\).

4. **Substitute and expand:**
   Substituting dY into the differential of f:
   $$ \begin{aligned} & df = \text{tr} \left( \left(\frac{\partial f}{\partial Y}\right)^\top (X^\top dX + (dX)^\top X) \right) \end{aligned} $$
   $$ \begin{aligned} & = \text{tr} \left( \frac{\partial f}{\partial Y} X^\top dX \right) + \text{tr} \left( \frac{\partial f}{\partial Y} (dX)^\top X \right) \end{aligned} $$
   Using the cyclic property of the trace operation, this expression simplifies to:
   $$ \begin{aligned} & 2 \text{tr} \left( \frac{\partial f}{\partial Y} X^\top dX \right) \end{aligned} $$

5. **Relate back to \( X \):**
   $$ \begin{aligned} & \frac{\partial f(X^\top X)}{\partial X} = 2X \frac{\partial f(X^\top X)}{\partial (X^\top X)} \end{aligned} $$
   which matches our formula to prove.

### Conclusion:
The chain rule in matrix calculus allows us to relate the differential of f in terms of X and \(X^\top X\), confirming the lemma. The factor of 2 arises from the derivative of a product, where each term contributes equally.

Construct the negative log-likelihood $\ell(\mathbf{U}, \mathbf{\Psi}) = -\log\mathcal{L}(\mathbf{X}; \mathbf{U}, \mathbf{\Psi})$, according to the FA model assumptions and show that its gradients with respect to the model parameters are 
$$
\nabla_{\mathbf{U}} \ell(\mathbf{U}, \mathbf{\Psi}) = 2 \mathbf{P}\mathbf{U}
$$

$$
\nabla_{\mathbf{\Psi}} \ell(\mathbf{U}, \mathbf{\Psi}) = \operatorname{diag}(\mathbf{P})

$$
\mathbf{P} = \nabla_{\mathbf{U}\mathbf{U}^\top + \mathbf{\Psi}} \ell(\mathbf{U}, \mathbf{\Psi}) = \frac12 \mathbf{Q} - \frac{1}{2n} \mathbf{Q} \mathbf{X}^\top \mathbf{X} \mathbf{Q}
$$
and
$$
\mathbf{Q} = (\mathbf{U}\mathbf{U}^\top + \mathbf{\Psi})^{-1}.

### Solution

#### Step 1: Define the Negative Log-Likelihood
The negative log-likelihood for the factor analysis model can be formulated as:
$$
\ell(\mathbf{U}, \mathbf{\Psi}) = -\log \det(\mathbf{U} \mathbf{U}^\top + \mathbf{\Psi}) + \operatorname{tr}\left((\mathbf{U} \mathbf{U}^\top + \mathbf{\Psi})^{-1} \mathbf{X} \mathbf{X}^\top\right)
$$
where \(\mathbf{X}\) is the data matrix, \(\mathbf{U}\) is the matrix of factor loadings, and \(\mathbf{\Psi}\) is the diagonal matrix of specific variances.

#### Step 2: Compute \(\mathbf{Q}\)
As given,
$$
\mathbf{Q} = (\mathbf{U}\mathbf{U}^\top + \mathbf{\Psi})^{-1}
$$
This is the precision matrix of the observed data.

#### Step 3: Derive Gradient with respect to \(\mathbf{U}\)

Applying the matrix calculus rules, we find the gradient of \(\ell\) with respect to \(\mathbf{U}\):
$$
\nabla_{\mathbf{U}} \ell(\mathbf{U}, \mathbf{\Psi}) = 2 \mathbf{Q} \mathbf{X} \mathbf{X}^\top \mathbf{U} - 2 \mathbf{Q} \mathbf{U}
$$
Simplifying, and assuming symmetrical properties,
$$
\nabla_{\mathbf{U}} \ell(\mathbf{U}, \mathbf{\Psi}) = 2 \mathbf{P} \mathbf{U}
$$
where \(\mathbf{P} = \frac{1}{2} \mathbf{Q} - \frac{1}{2n} \mathbf{Q} \mathbf{X}^\top \mathbf{X} \mathbf{Q}\).

#### Step 4: Derive Gradient with respect to \(\mathbf{\Psi}\)
The gradient with respect to \(\mathbf{\Psi}\) focuses on the diagonal elements:
$$
\nabla_{\mathbf{\Psi}} \ell(\mathbf{U}, \mathbf{\Psi}) = \operatorname{diag}(\mathbf{Q} - \mathbf{Q} \mathbf{X} \mathbf{X}^\top \mathbf{Q})
$$
Simplifying to:
$$
\nabla_{\mathbf{\Psi}} \ell(\mathbf{U}, \mathbf{\Psi}) = \operatorname{diag}(\mathbf{P})
$$

### Conclusion
These gradients are essential for implementing gradient descent in optimization algorithms used to find the parameters \(\mathbf{U}\) and \(\mathbf{\Psi}\) that minimize the negative log-likelihood, providing an efficient way to perform parameter estimation in factor analysis models.
