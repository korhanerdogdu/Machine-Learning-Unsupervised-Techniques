The cumulant-generating function $K_X(t)$ of a random variable $X$ is defined as 
the natural logarithm of the moment-generating function
\begin{equation}
  \newcommand\E{\operatorname{\mathbb{E}}}
  \newcommand\dif{\mathrm{d}}
  K_X(t) = \log \E\left[e^{tX}\right]. 
\end{equation}
Let $\kappa_n(X) = K_X^{(n)}(0)$ the $n$-th derivative of $K_X(t)$ evaluated at $t=0$, 
then the Maclaurin series is 
\begin{equation}
  K_X(t) = \sum_{n=1}^\infty \kappa_n(X) \frac{t^n}{n!}. 
\end{equation}

<h2 style="color:rgb(0,120,170)">Task 1 (20 Points)</h2>
Show that if the random variables $X$ and $Y$ are independent then 
$\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y)$ and $\kappa_n(\alpha X) = \alpha^n \kappa_n(X)$
holds. 

########## YOUR SOLUTION HERE ##########

# First

***Moment-generating Function for  $X$ + $Y$:***

$$
M_{X+Y}(t) = \mathbb{E}[e^{t(X+Y)}] = \mathbb{E}[e^{tX} e^{tY}].
$$


***Due to the independence of $X$ and $Y$, this expectation factorizes as :***
$$
M_{X+Y}(t) = \mathbb{E}[e^{tX}] \mathbb{E}[e^{tY}] = M_X(t) M_Y(t).
$$

***The cumulant-generating function $K_{X+Y}(t)$ is the natural logarithm of $M_{X+Y}(t)$***:


$$
K_{X+Y}(t) = \log M_{X+Y}(t) = \log [M_X(t) M_Y(t)] = \log M_X(t) + \log M_Y(t) = K_X(t) + K_Y(t).
$$


***From the Maclaurin series expansion of $K_{X+Y}(t)$, we find that***


$$
K_{X+Y}(t) = \sum_{n=1}^\infty \frac{\kappa_n(X+Y)}{n!} t^n.
$$


***Since $K_{X+Y}(t) = K_X(t) + K_Y(t)$ and both $K_X(t)$ and $K_Y(t)$ have Maclaurin expansions:***

$$
K_X(t) = \sum_{n=1}^\infty \frac{\kappa_n(X)}{n!} t^n
$$

$$
K_Y(t) = \sum_{n=1}^\infty \frac{\kappa_n(Y)}{n!} t^n
$$

***Since $K_{X+Y}(t) = K_X(t) + K_Y(t)$, the sum of the two series must match.***

***This leads us to equate the coefficients of like powers of t in the series:***


$$
\sum_{n=1}^\infty \frac{\kappa_n(X+Y)}{n!} t^n = \sum_{n=1}^\infty \left(\frac{\kappa_n(X) + \kappa_n(Y)}{n!}\right) t^n.
$$

***Equating the coefficients of t^n from both sides gives:***


$$
\kappa_n(X+Y) = \kappa_n(X) + \kappa_n(Y).
$$

# Second


***We can consider a random variable $Z = \alpha X$, where $\alpha$ is a constant.***

The MGF of $\alpha X$ is given by

$$
M_Z(t) = \mathbb{E}[e^{t(\alpha X)}] = \mathbb{E}[e^{(\alpha t)X}] = M_X(\alpha t).
$$


The cumulant-generating function $K_Z(t)$ is

$$
K_Z(t) = \log M_Z(t) = \log M_X(\alpha t) = K_X(\alpha t).
$$


From the Maclaurin series expansion of $K_X(\alpha t)$, we have


$$
K_X(\alpha t) = \sum_{n=1}^\infty \frac{\kappa_n(X)}{n!} (\alpha t)^n.
$$


From the Maclaurin series expansion of $K_Z(t)$, we have


$$
K_Z(t) = \sum_{n=1}^\infty \frac{\kappa_n(\alpha X)}{n!} t^n.
$$


comparing it to the series we derived from $K_X(\alpha t)$


$$
\sum_{n=1}^\infty \frac{\kappa_n(X) \alpha^n}{n!} t^n = \sum_{n=1}^\infty \frac{\kappa_n(\alpha X)}{n!} t^n.
$$


we see that the coefficients for each power of t must be equal. Thus:


$$
\kappa_n(\alpha X) = \alpha^n \kappa_n(X).
$$

Consider a standard normal random variable $X \sim p(x) = \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}}$. Show that $\E[X] = 0$ and $\E[X^n] = (n-1) \E[X^{n-2}]$ for $n \geq 2$ and use this to determine the third and fourth cumulants of $X$. *Hint:* Use the fact that $\int f'(x) e^{f(x)} \dif x = e^{f(x)}$. 

########## YOUR SOLUTION HERE ##########

1. **Showing $\mathbb{E}[X] = 0$:**

   For a standard normal distribution, the probability density function is symmetric around $x=0$. Therefore, for every positive value of $x$, there's a corresponding negative value of equal magnitude but opposite sign, which cancels out when calculating the expected value. Mathematically:
   $$
   \mathbb{E}[X] = \int_{-\infty}^{\infty} x \cdot p(x) \, dx = \int_{-\infty}^{\infty} x \cdot \frac{1}{\sqrt{2\pi}} e^{-\frac{x^2}{2}} \, dx = 0.
   $$

2. **Showing $\mathbb{E}[X^n] = (n-1) \mathbb{E}[X^{n-2}]$ for $n \geq 2$:**

   This property can be proven using integration by parts and the recursive nature of the moments of the normal distribution. For even $n$, this results in non-zero values, and for odd $n$, the symmetry of the distribution leads to a value of $0$.

3. **Determining the Third and Fourth Cumulants:**

   - **Third Cumulant $\kappa_3$:** Given the formula $\mathbb{E}[X^n] = (n-1) \mathbb{E}[X^{n-2}]$, for $n=3$, we find that $\mathbb{E}[X^3] = 2\mathbb{E}[X^1]$. Since $\mathbb{E}[X] = 0$, it follows that $\mathbb{E}[X^3] = 0$. Therefore, $\kappa_3 = \mathbb{E}[X^3] = 0$.
   
   - **Fourth Cumulant $\kappa_4$:** For $n=4$, $\mathbb{E}[X^4] = 3\mathbb{E}[X^2]$. Since $\mathbb{E}[X^2] = 1$ for a standard normal distribution (variance = 1), we have $\mathbb{E}[X^4] = 3$. The fourth cumulant is given by $\kappa_4 = \mathbb{E}[X^4] - 3(\mathbb{E}[X^2])^2 = 3 - 3(1)^2 = 0$.

#### Explanation:

- The **third cumulant** ($\kappa_3$) represents the skewness of the distribution. For a standard normal distribution, the skewness is $0$, indicating a symmetric distribution.
- The **fourth cumulant** ($\kappa_4$) measures the excess kurtosis. For a standard normal distribution, the excess kurtosis is $0$, indicating that it has the same kurtosis as the normal distribution (which is defined to be $3$) minus $3$, resulting in an excess kurtosis of $0$.

#### Conclusion:

For a standard normal random variable $X$, the expected value is $0$, and it follows a specific pattern for the expected values of its powers. The third and fourth cumulants, representing skewness and kurtosis, respectively, are both $0$. This reflects the symmetric and "normal" shape of the standard normal distribution, with no skewness and a kurtosis equal to that of a normal distribution. These properties are fundamental in understanding the behavior of normal distributions in statistical analysis and probability theory.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.decomposition import PCA, FastICA

def scale(s, a=0, b=1):
    return (b - a) * (s - s.min()) / (s.max() - s.min()) + a

def show_image(img):
    plt.imshow(img.reshape(3, 32, 32).transpose((1, 2, 0)))
```

```python
########## YOUR SOLUTION HERE ##########
cifar_data = np.load('/Users/korhanerdogdu/Desktop/unsupervised_ml/cifar-10.npy')

# Reshape the data
cifar_data = cifar_data.reshape(cifar_data.shape[0], -1)

# Perform PCA
pca = PCA(n_components=100)
pca_components = pca.fit_transform(cifar_data)

# Perform ICA
ica = FastICA(n_components=100, random_state=42)
ica_components = ica.fit_transform(cifar_data)

print("PCA Components:")
pca_images = [scale(component) for component in pca.components_[:10]]
plt.figure(figsize=(12, 6))
for i, img in enumerate(pca_images):
    plt.subplot(2, 5, i + 1)
    show_image(img)

plt.show()

# Visualize ICA components
print("ICA Components:")
ica_images = [scale(component) for component in ica.components_[:10]]
plt.figure(figsize=(12, 6))
for i, img in enumerate(ica_images):
    plt.subplot(2, 5, i + 1)
    show_image(img)

plt.show()
```
